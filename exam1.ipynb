```python
def max_intnum(*p) :
    '''
    p : 숫자들을 1개 이상 전달
    전달된 숫자들중에서 정수만 추출하여 최댓값을 리턴
    '''
    listv = [ e for e in p if e.is_integer() ]
    print(listv)
    return max(listv)
```

## 파이썬의 네임스페스(Namespace)

- 네임스페이스란 이름(객체)을 통해 특정 객체(값)를 구분하고 참조할 수 있는 범위이다.
- 변수, 함수, 클래스 등 모든 파이썬 객체는 자신만의 네임스페이스를 가진다.
- 파이썬에서는 여러 종류의 네임스페이스(글로벌, 로컬, 클래스 등)가 존재한다.
- 각 네임스페이스는 서로 독립적으로 관리된다. 



### 파이썬은 실행되는 컨텍스트에 따라 다양한 네임스페이스를 사용한다.
- **내장 네임스페이스(Built-in Namespace)**:<br>
  print(), len() 등 파이썬을 실행하는 데 필요한 기본 함수들이 포함된 공간 
- **전역 네임스페이스(Global Namespace)**:<br>
  특정 모듈이나 스크립트 파일 내에서 정의된 변수나 함수가 속하는 네임스페이스 
- **지역 네임스페이스(Local Namespace)**:<br>
  함수가 호출될 때 생성되는 임시적인 네임스페이스로, 해당 함수 안에서 정의된 변수들을 저장 
- **클래스 네임스페이스(Class Namespace)**:<br>
  클래스가 정의될 때 생성되며, 해당 클래스의 클래스 변수나 메서드가 저장되는 공간

### dir()

- dir() 함수는 현재 스코프 또는 전달된 객체 영역에 연결된 네임스페이스에 있는 모든 이름을 출력한다.
- 아규먼트 없이 dir()을 호출하면 현재 로컬 범위에 정의된 변수, 함수 또는 import 된 모듈 등이 표시된다. 
- dir(아규먼트) 함수를 사용하여 특정 객체(클래스, 인스턴스, 모듈 등)에 대한 정보(변수와 메서드 목록)를 얻을 수 있다.
- dir(MyClass)를 호출하면 MyClass 클래스의 멤버(속성, 메서드)를 확인할 수 있다.

```python
# 부모 클래스를 선언합니다.
class Parent(object):
    def __init__(self):
        self.value = "부모의 test 변수"
        print("Parent 클래스의 __init()__ 메서드가 호출되었습니다.")

    def test(self):
        print("Parent 클래스의 test() 메서드입니다.")

# 자식 클래스를 선언합니다.
class Child(Parent):
    def __init__(self):
        super().__init__()
        print("Child 클래스의 __init()__ 메서드가 호출되었습니다.")
```

```python
# 자식 클래스의 인스턴스를 생성하고 부모의 메서드 호출
obj = Child()
```

```python
class Human:
    def __init__(self, age, name):
        self.age = age
        self.name = name

    def intro(self):
        print(f"{self.age}세 {self.name}입니다.")

    def eat(self):
        print("밥을 먹는다")
```

```python
class Student(Human):
    def __init__(self, age, name, stunum):
        super().__init__(age, name)
        self.stunum = stunum

    def intro(self):   # 메서드 오버라이딩
        super().intro()
        print(f"학번 : {self.stunum}")

    def study(self):   # 메서드 추가
        print("파이썬을 공부해요")
```

```python
class Human(object):
    def __init__(self, hname):
        self.hname = hname

    def info_human(self):
        print(f"이름은 {self.hname} 입니다. {self.book_name}에서 만날 수 있어요.")

class Parent(Human):
    def __init__(self, pname):
        super().__init__("이득촌")
        self.pname = pname

    def info_parent(self):
        print(f"이름은 {self.pname} 입니다.  {self.book_name}에서 만날 수 있어요.")

class Child(Parent):
    def __init__(self, name, book_name):
        super().__init__("이말룡")
        self.name = name
        self.book_name = book_name

    def info_child(self):
        print(f"이름은 {self.name} 입니다. {self.book_name}에서 만날 수 있어요.")
```

```python
actor.info_child()
```

### 클래스 변수와 클래스 메서드


```python
class Student:
    # 클래스 변수
    count = 0
    students = []

    # 클래스 메서드
    @classmethod
    def print(cls):
        print("cls에 전달된 값 :",cls)
        print("------------- 학생 목록 -------------")
        print("이름\t\t총점\t\t평균")
        for student in cls.students:
            print(str(student))
        print("------------------------------------")

    # 초기화 메서드
    def __init__(self, name, korean, math, english, science):
        self.name = name
        self.korean = korean
        self.math = math
        self.english = english
        self.science = science
        Student.count += 1
        Student.students.append(self)

    # 인스턴스 메서드
    def get_sum(self):
        return self.korean + self.math + \
               self.english + self.science

    # 인스턴스 메서드
    def get_average(self):
        return self.get_sum() / 4

    # 인스턴스 메서드 : 오버라이딩
    def __str__(self):
        return f"{self.name}\t\t{self.get_sum()}\t\t{self.get_average()}"
```

```python
class Member:
    def __init__(self, email, password):
        self.email = email
        self.password = password

    @classmethod
    def fromTuple(cls, tup):
        return cls(tup[0], tup[1])

    @classmethod
    def fromDictionary(cls, dic):
        return cls(dic["email"], dic["password"])

    def __str__(self):
        return f"{self.email} ----- {self.password}"
```

### 정적(static) 메서드
- @staticmethod 데코레이터를 사용해서 클래스에 메서드를 선언하면 해당 메서드는 정적(static) 메서드가 된다. 
- 정적 메서드는 인스턴스 메서드나 클래스 메서드와 달리 첫번째 매개 변수가 할당되지 않는다.
- 정적 메서드 내에서는 인스턴스/클래스 속성에 접근하거나, 인스턴스/클래스 메서드를 호출하는 것이 불가능하다.
- 정적 메서드는 유틸리티 메서드를 구현할 때 많이 사용된다.

```python
class StringUtils:
    @staticmethod
    def toCamelcase(text):
        words = iter(text.split("_"))
        return next(words) + "".join(i.title() for i in words)

    @staticmethod
    def toSnakecase(text):
        letters = ["_" + i.lower() if i.isupper() else i for i in text]
        return "".join(letters).lstrip("_")
```

```python
print(StringUtils.toCamelcase("last_modified_date"))
print(StringUtils.toSnakecase("lastModifiedDate"))
```

위의 두 개의 메서드는 매개 변수로 넘어온 문자열에만 의존하는 순수한(pure) 함수이기 때문에 굳이 클래스의 일부로 선언할 필요는 없지만
이렇게 비슷한 류의 여러 유틸리티 메서드를 하나의 클래스의 묶어두고 싶을 때 정적 메서드로 선언할 수 있다.

## 데코레이터 

- 데코레이터는 기존 코드를 수정하지 않고도 함수나 메서드의 기능을 확장하거나 변경할 수 있게 해주는 도구이다.
- 데코레이터는 함수를 감싸는 또 다른 함수이다.
- 다른 함수의 동작을 수정하거나 확장하는 데 사용할 수 있는 특별한 유형의 함수이다.
- 코드를 직접 변경하지 않고도 추가 기능으로 함수를 래핑하거나 장식할 수 있다.
- 무분별한 데코레이터의 사용은 코드의 복잡도를 높일 수 있다.

### 데코레이터의 정의와 사용
정의된 데코레이터를 사용하는 함수나 메서드 앞에 '@' 기호와 데코레이터 이름을 지정한다.

```python
# decorator를 정의하는 함수. 인자로 함수를 받는다.
def my_decorator(func):
    def wrapper():
        print("함수 실행 전 무언가를 합니다.")
        func() # 인자로 받은 함수를 실행한다.
        print("함수 실행 후 무언가를 합니다.")
    return wrapper

# @ + decorator 함수명을 앞에 쓰는 사용할 수 있다.
@my_decorator
def say_hello():
    print("Hello!")  
    
say_hello()
```

```python
# 어노테이션 없이 데코레이터 효과
def say_hello2():
    print("Hello2!")

deco_say_hello2 = my_decorator(say_hello2)
deco_say_hello2()
```

```python
def security_system(house):
    def protected_house():
        print("보안 시스템 활성화")
        house()
        print("보안 시스템 비활성화")
    return protected_house
```

```python
# 어노테이션으로 데코레이션 효과
@security_system
def my_house():
    print("이것은 내 집입니다.")
```

```python
import time
import logging

# 실행 시간을 로깅하는 decorator
def log_decorator(func):
    def wrapper():
        start_time = time.time()
        result = func()
        end_time = time.time()
        logging.warning(f'{func.__name__} 실행 시간: {end_time - start_time} 초') # logging 코드 추가
        return result
    return wrapper

@log_decorator
def my_function():
    time.sleep(2)
```

데코레이터를 정의할 때 *args와 **kargs를 사용하면 원래 함수가 받는 인자들을 그대로 전달할 수 있어 어떤 형태의 함수든 데코레이터를 적용할 수 있다.

```python
def test_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"{func.__name__} 호출됨: args={args}, kwargs={kwargs}")
        return func(*args, **kwargs)
    return wrapper
 
@test_decorator
def add(a, b):
    return a + b
 
print(add(3, 5))
```

### 데코레이터를 사용하면 얻을 수 있는 이점

- 관심사의 분리: 핵심 로직과 부가 기능을 분리할 수 있다.
- 가독성 향상: 복잡한 로직을 간결하게 표현할 수 있다.
- 유지보수 용이성: 기능 추가나 변경이 쉬워진다.

### 데코레이터 활용에 기반이 되는 파이썬 구문

- 파이썬에서 함수는 '일급 객체'<br>
  이는 함수를 변수에 할당하거나, 다른 함수의 인자로 전달하거나, 함수에서 반환할 수 있다는 의미이다.
- 중첩 함수<br>
  중첩 함수는 다른 함수 내부에 정의된 함수이다. 이는 데코레이터의 구조를 이해하는 데 중요하다.
- 클로저<br>
  **클로저는 자신을 둘러싼 스코프의 변수를 기억하는 함수**이다. 이는 데코레이터가 작동하는 핵심 메커니즘이다.

```python
def greet(name):
    return f"Hello, {name}!"

# 함수를 변수에 할당
say_hi = greet

# 함수를 인자로 전달
def apply_func(func, value):
    return func(value)

result = apply_func(say_hi, "Alice")
print(result)  
```

```python
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function

add_five = outer_function(5)
result = add_five(3)
print(result) 
```

#### 여러개의 데코레이터 사용하기

```python
def star(func):
    def wrapper(*args, **kwargs):
        print("*" * 30)
        func(*args, **kwargs)
        print("*" * 30)
    return wrapper

def percent(func):
    def wrapper(*args, **kwargs):
        print("%" * 30)
        func(*args, **kwargs)
        print("%" * 30)
    return wrapper

@star
@percent
def printer(msg):
    print(msg)

printer("Hello")
```

#### 클래스에 데코레이터 적용기기

```python
def singleton(cls):
    instances = {}
    def get_instance(*args, **kwargs):
        if cls not in instances:
            instances[cls] = cls(*args, **kwargs)
        return instances[cls]
    return get_instance

@singleton
class DatabaseConnection:
    def __init__(self):
        print("Initializing database connection")

# 항상 같은 인스턴스를 반환
db1 = DatabaseConnection()
db2 = DatabaseConnection()
print(db1 is db2) 
print(db1 is DatabaseConnection()) 
```
